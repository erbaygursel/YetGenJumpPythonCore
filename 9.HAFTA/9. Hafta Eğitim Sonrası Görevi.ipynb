# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import random
import time

class Kumanda:
    
    def __init__(self):
        self.acik = False
        self.ses = 50
        self.kanallar = ['TRT', 'CNN', 'FOX', 'NTV']
        self.secili_kanal = 'TRT'
        
    def tv_ac(self):
        if self.acik:
            print("Televizyon zaten açık.")
        else:
            self.acik = True
            print("Televizyon açıldı.")
            
    def tv_kapat(self):
        if self.acik:
            self.acik = False
            print("Televizyon kapandı.")
        else:
            print("Televizyon zaten kapalı.")
            
    def ses_ayarla(self):
        if self.acik:
            yeni_ses = input("Ses seviyesini girin (0-100): ")
            self.ses = int(yeni_ses)
            print("Ses seviyesi:", self.ses)
        else:
            print("Televizyon kapalı. Ses ayarlanamaz.")
            
    def kanal_ekle(self, kanal):
        if self.acik:
            self.kanallar.append(kanal)
            print(kanal, "kanal listesine eklendi.")
        else:
            print("Televizyon kapalı. Kanal eklenemez.")
            
    def kanal_sayisi(self):
        print("Toplam", len(self.kanallar), "kanal var.")
        
    def rastgele_kanal(self):
        if self.acik:
            rastgele_index = random.randint(0, len(self.kanallar)-1)
            self.secili_kanal = self.kanallar[rastgele_index]
            print("Şu anki kanal:", self.secili_kanal)
        else:
            print("Televizyon kapalı. Rastgele kanala geçilemez.")
            
    def tv_bilgileri(self):
        print("Televizyon durumu:", "Açık" if self.acik else "Kapalı")
        print("Ses seviyesi:", self.ses)
        print("Kanallar:", self.kanallar)
        print("Şu anki kanal:", self.secili_kanal)
        
kumanda = Kumanda()

while True:
    islem = input("Hangi işlemi yapmak istiyorsunuz? (q = çıkış): ")
    
    if islem == "q":
        print("Programdan çıkılıyor...")
        break
    
    elif islem == "1":
        kumanda.tv_ac()
        
    elif islem == "2":
        kumanda.tv_kapat()
        
    elif islem == "3":
        kumanda.ses_ayarla()
        
    elif islem == "4":
        kanal_isimleri = input("Kanal isimlerini virgülle ayırarak girin: ")
        kanal_listesi = kanal_isimleri.split(",")
        
        for kanal in kanal_listesi:
            kumanda.kanal_ekle(kanal.strip())
        
    elif islem == "5":
        kumanda.kanal_sayisi()
    elif islem == "6":
        kumanda.rastgele_kanal()
    elif islem == "7":
        kumanda.tv_bilgileri()
    else:
        print("Geçersiz işlem.")
```
