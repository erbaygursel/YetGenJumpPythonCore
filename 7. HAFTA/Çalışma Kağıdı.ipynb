# Problem 3 Virgül ile Ayırma

Kullanıcıdan bir metin halinde virgül ile ayrılmış sayılar alınız ve aldığınız her bir sayıyı listeye atayınız, daha sonra listedeki her elemanı toplayıp ekrana bastıran bir program yazınız.

<h4>Sample Input</h4>
<p>
Lutfen sayilari virgul ile ayrirarak giriniz: 5,14,7,2,1
</p>
<h4>Output</h4>

Toplam: 29

```python
# 3.Problemin Çözümünü Buraya Yazınız
adet = int(input("Kaç tane sayı gireceksiniz?"))
sayıList = []
toplam = 0
for i in range(adet):
    sayılar = int(input("Hangi sayıları yazmak isterseniz?"))
    sayıList.append(sayılar)
    toplam += sayılar
print(f"Seçtiğiniz sayılar: {sayıList} ve toplamları: {toplam}")
```

# Problem 4 Büyük Sayı

# Problem 5 Ürün Kayıt Uygulaması
Programımız ilk olarak kullanıcıdan, kaydı yapılacak ürünlerin sayısını isteyecek. Ardından yazılmış olan ürün sayısı kadar ürün ismi ve ürün fiyatı bilgisini kullanıcıdan alacak. Son olarak ise ürünleri ekrana yazdıracağız.

<h4>Sample Input:</h4>
<p>
Lutfen kac adet urun gireceginizi yaziniz: 3
<br/>
1. Urununuzun ismini giriniz: Mentos
<br/>
1. Urununuzun fiyatini giriniz: 1
<br/>
2. Urununuzun ismini giriniz: Meybuz
<br/>
2. Urununuzun fiyatini giriniz: 0.25
<br/>
3. Urununuzun ismini giriniz: Uzun Şeker
<br/>
3. Urununuzun fiyatini giriniz: 0.75
<br/>
<h4>Output:</h4>
1. Urun: Mentos - 1 TL
<br/>
2. Urun: Meybuz - 0.25 TL
<br/>
3. Urun: Uzun Şeker - 0.75 TL
</p>

```python
# 5.Problemin Çözümünü Buraya Yazınız
urun_sayisi = int(input("Lütfen kaç adet ürün kaydı yapılacağını yazınız: "))

urunler = []

for i in range(urun_sayisi):
    urun_adi = input("Ürününüzün ismini giriniz: ")
    urun_fiyati = float(input("Ürününüzün fiyatını giriniz: "))
    urunler.append((urun_adi, urun_fiyati))

for i, urun in enumerate(urunler):
    print(f"{i+1}. Urun: {urun[0]} - {urun[1]} TL")

```

# Problem 6 Türkçe Karakter Çevirici
Programa girilen dosya ismindeki Türkçe karakterleri, İngilizce'ye çeviren programı yazınız.

<h4>Sample Input:</h4>
<p>
Lutfen dosya ismi giriniz: fotoğraf.jpeg
<br>
Dosya ismi: fotograf.jpeg
<br>
<br>
Lutfen dosya ismi giriniz: ışık.png
<br>
Dosya ismi: isik.png
</p>

```python
# 6.Problemin Çözümünü Buraya Yazınız
def convert_turkish_chars(file_name):
    türkçe_harf = {
        'ç': 'c',
        'Ç': 'C',
        'ğ': 'g',
        'Ğ': 'G',
        'ı': 'i',
        'İ': 'I',
        'ö': 'o',
        'Ö': 'O',
        'ş': 's',
        'Ş': 'S',
        'ü': 'u',
        'Ü': 'U',
    }

    converted_name = ''
    for harf in file_name:
        if harf in türkçe_harf:
            converted_name += türkçe_harf[harf]
        else:
            converted_name += harf

    return converted_name


file_name = input("Lutfen dosya ismi giriniz: ")
converted_name = convert_turkish_chars(file_name)
print("Dosya ismi:", converted_name)
```

# Problem 7 Sayı Tahmin Oyunu ama Bilgisayar Tahmin Ediyor :)
Sayı tahmin oyunu: İki oyuncudan birinin, 0 ile 100 arasında bir sayı tutması ile başlar. Diğer oyuncu ise tahminler yaparak tutulan sayıyı bulmaya çalışır ve yaptığı her tahminde sayıyı tutmuş olan oyuncudan YUKARI veya AŞAĞI olarak geri bildirim alır. Sayıyı bildiğinde ise rakip oyuncu kaç tahminde (kaç denemede) bildiğini söyler.

Senden yazmanı beklediğimiz program ise Python'ın tahminleri yapan tarafta olacağı bir oyun oluşturman. Yani sen 0 ile 100 arasında bir sayı tutacaksın ardından program senin tuttuğunuz sayıyı bulmaya çalışacak.
<h4>Sample Input:</h4>
Bu örnekte aklınızdan 90 tuttuğunuzu varsayalım.
<p>
Oyun başlıyor lutfen aklindan bir sayi tut! Tuttun mu, hazır mısın? : 1
<br>
Tahmin: 50
<br>
Durum: Yukari
<br>
Tahmin: 75
<br>
Durum: Yukari
<br>
Tahmin: 87
<br>
Durum: Yukari
<br>
Tahmin: 93
<br>
Durum: Asagi
<br>
Tahmin: 90
<br>
Durum: Bildin
</p>

```python
# 7.Problemin Çözümünü Buraya Yazınız
import random

def sayıTahmin():
    altLimit = 0
    ustLimit = 100
    count = 0

    print("Oyun başlıyor! Bir sayı tutun ve hazır olduğunuzda 1'e basın.")

    ready = input()
    if ready != '1':
        return

    while True:
        count += 1
        tahmin = random.randint(altLimit, ustLimit)
        print("Tahmin:", tahmin)

        feedback = input("Durum: ")

        if feedback.lower() == "bildin":
            print("Tebrikler! Sayıyı", count, "tahminde buldum.")
            break
        elif feedback.lower() == "Yukarı":
            altLimit = tahmin + 1
        elif feedback.lower() == "Aşağı":
            ustLimit = tahmin - 1
        else:
            print("Geçersiz geri bildirim. 'Yukari', 'Asagi' veya 'Bildin' şeklinde yanıt verin.")


sayıTahmin()
```

# Problem 9 Tek-Çift

Kullanıcıdan 'q' tuşuna basana kadar rastgele sayılar alın. Tek olanları bir listeye, çift olanları bir başka listeye büyükten küçüğe sıralanacak şekilde atın. Listeleri ekrana yazdıran bir kod yazın.

```python
# 9.Problemin Çözümünü Buraya Yazınız
def main():
    tekSayılar = []
    ciftSayılar = []

    while True:
        number = input("Bir sayı girin (Çıkmak için 'q' tuşuna basın): ")
        if number == 'q':
            break

        try:
            number = int(number)
            if number % 2 == 0:
                ciftSayılar.append(number)
            else:
                tekSayılar.append(number)
        except ValueError:
            print("Geçersiz bir giriş yaptınız. Lütfen bir sayı girin.")

    tekSayılar.sort(reverse=True)
    ciftSayılar.sort(reverse=True)

    print("Tek Sayılar (Büyükten Küçüğe):", tekSayılar)
    print("Çift Sayılar (Büyükten Küçüğe):", ciftSayılar)


main()
```
