```python
import openai
import pyttsx3
import speech_recognition as sr
import tkinter as tk

# OpenAI API anahtarını ayarlayın
openai.api_key = "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"

# metin-konuşma motorunu başlatın
engine = pyttsx3.init()

# konuşma tanıma işlemini ayarlayın
r = sr.Recognizer()
mic = sr.Microphone(device_index=1)

# Tkinter penceresi oluşturun
window = tk.Tk()
window.title("OpenAI Bot")
window.geometry("400x400")

# metin kutularını oluşturun
user_text = tk.Text(window, height=5, width=50)
user_text.pack()
bot_text = tk.Text(window, height=5, width=50)
bot_text.pack()

# düğmeleri oluşturun
continue_button = tk.Button(window, text="Konuşmaya devam et")
continue_button.pack()
exit_button = tk.Button(window, text="Çıkış")
exit_button.pack()

conversation = ""
user_name = "You"
bot_name = "DAI"

def continue_conversation():
    global conversation
    with mic as source:
        print("\nlistening...")
        r.adjust_for_ambient_noise(source, duration=0.2)
        audio = r.listen(source)
    print("no longer listening.\n")

    try:
        user_input = r.recognize_google(audio, language='tr-TR')
        print(user_input)
        if user_input.lower() in ["çıkış", "exit"]:
            window.destroy()
            return
    except:
        user_input = ""

    prompt = user_name + ": " + user_input + "\n" + bot_name+ ": "

    conversation += prompt  # allows for context

    # fetch response from OpenAI API
    response = openai.Completion.create(
        engine='text-davinci-002',
        prompt=conversation,
        max_tokens=100,
        n=1,
        stop=None,
        temperature=0.7,
    )

    response_str = response["choices"][0]["text"].replace("\n", "")
    response_str = response_str.split(user_name + ": ", 1)[0].split(bot_name + ": ", 1)[0]

    conversation += response_str + "\n"
    print(response_str)

    bot_text.insert(tk.END, bot_name + ": " + response_str + "\n")
    engine.say(response_str)
    engine.runAndWait()

continue_button.config(command=continue_conversation)

def exit_conversation():
    window.destroy()

exit_button.config(command=exit_conversation)

window.mainloop()
```
